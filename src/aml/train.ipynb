Denne fil skal bruges til at træne vores maskinlæringsmodel.
Når vi har gjort data klar (X og y), kan vi vælge en model og lade den "lære" af data.

Vi kan starte med en simpel model (fx lineær regression) og senere prøve andre.
Vi kan også tilføje ting som hyperparametre, gemme modellen, eller sammenligne flere modeller.
