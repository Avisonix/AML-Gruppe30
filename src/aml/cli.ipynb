Denne fil er "startpunktet" for vores projekt.
CLI betyder "Command Line Interface" — altså noget man kan køre fra terminalen.

Når vi senere vil teste hele jvores kode, kan vi skrive:
- python -m aml.cli

Denne fil skal på sigt samle hele jeres pipeline:
1. Hent data
2. Rens data
3. Lav features
4. Træn model
5. Evaluer model

Eller lignede afhænging af hvad projektet præcist ender med at indeholde.
