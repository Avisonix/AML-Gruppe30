Denne fil skal bruges til at vælge og/eller skabe de "features" (variabler), som vores maskinlæringsmodel skal bruge.

Her kan vi fx:
- vælge bestemte kolonner, der er relevante
- lave nye kolonner ud fra eksisterende data (fx alder = 2025 - fødselsår)
- konvertere tekst (kategorier) til tal (one-hot encoding)
- splitte data i X (input) og y (det vi vil forudsige)

Det vi laver her, er det data, som vores model skal trænes på.
